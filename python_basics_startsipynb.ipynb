```python
colors = ["red", "green", "blue"]

print("\n--- Using enumerate() ---")
for index, color in enumerate(colors):
    print(f"Color #{index + 1} is {color}.")
# Output:
# Color #1 is red.
# Color #2 is green.
# Color #3 is blue.
```

```python
numbers=[1,2,3,0,5,6]
print('\n---using break and countine---')
for num in numbers:
  if num==0:
    print("Found a zero,skipping this iteration(countinue),")
    continue
  if num>5:
    print(f"number {num} is too high,breaking the loop")
    break

  print(f"current number:{num}")

```
